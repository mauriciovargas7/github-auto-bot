# FAZ DOWNLOAD DE TODOS REPOSITÓRIOS DO GIT

```python
import requests
import os

"""
    Ao executar vai fazer o download de todos os repositórios que existirem no 
    github que estiver na variável "USUARIO" 

    Demonstração:

    GIT-HUB/
    └── repositórios/ 
        ├── 🤖github-auto-bot.ipynb
        ├── 📁repo1
        ├── 📁repo2
        ├── 📁repo3

"""


USUARIO = "NOME DO SEU USUÁRIO"
PASTA_DESTINO = os.getcwd() # Coloque esse arquivo (script) "github-auto-bot.ipynb" SOLTO NA pasta que vai ser seu GIT clone

os.makedirs(PASTA_DESTINO, exist_ok=True)

url = f"https://api.github.com/users/{USUARIO}/repos"

repos = requests.get(url).json()

for repo in repos:
    nome = repo["name"]
    clone_url = repo["clone_url"]

    caminho = os.path.join(PASTA_DESTINO, nome)

    if not os.path.exists(caminho):
        print(f"Clonando {nome}...")
        os.system(f'git clone {clone_url} "{caminho}"')
    else:
        print(f"{nome} já existe, pulando...")

print("✔ Tudo sincronizado!")
```

# FAZ UPLOAD DE TODOS REPOSITÓRIOS PRO GIT

```python
import os

"""
    Ao executar vai fazer o UPLOAD de todos os repositórios que existirem no 
    seu github, cadastrado no Git BASH

    Demonstração:

    GIT-HUB/
    └── repositórios/ 
        ├── 🤖github-auto-bot.ipynb
        ├── 📁repo1
        ├── 📁repo2
        ├── 📁repo3

"""

PASTA_DESTINO = os.getcwd() # Coloque esse arquivo (script) "github-auto-bot.ipynb" SOLTO NA pasta que vai ser seu GIT clone

for repo in os.listdir(PASTA_DESTINO):
    caminho = os.path.join(PASTA_DESTINO, repo)

    if os.path.isdir(os.path.join(caminho, ".git")):
        print(f"📤 Enviando {repo}...")

        os.system(f'cd "{caminho}" && git add .')
        os.system(f'cd "{caminho}" && git commit -m "Auto update"')
        os.system(f'cd "{caminho}" && git push')

    else:
        print(f"❌ {repo} não é um repositório git")

print("✔ Todos os repositórios enviados!")
```
